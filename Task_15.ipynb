```python
import numpy as np
import matplotlib.pyplot as plt
 
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))
 
sales = [10, 30, 12, 11]
date= [1, 5, 21, 22]
 
br1 = np.arange(len(sales))
br2 = [x + barWidth for x in br1]

plt.bar(br1, sales, color ='r', width = barWidth,
        edgecolor ='grey', label ='sales')
plt.bar(br2, date, color ='y', width = barWidth,
        edgecolor ='grey', label ='date')
 
plt.xlabel('category', fontweight ='bold', fontsize = 15)
plt.ylabel('data', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(sales))],
        ['baju', 'bajuu', 'topi', 'aksesoris'])

plt.title('SALES_DATA')
plt.legend()
plt.show()
```
