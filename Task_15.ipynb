```python
import numpy as np
import matplotlib.pyplot as plt
 
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))
 
sales = [10, 30, 12, 11]
date= [1, 5, 21, 22]
 
br1 = np.arange(len(sales))
br2 = [x + barWidth for x in br1]

plt.bar(br1, sales, color ='r', width = barWidth,
        edgecolor ='grey', label ='sales')
plt.bar(br2, date, color ='y', width = barWidth,
        edgecolor ='grey', label ='date')
 
plt.xlabel('category', fontweight ='bold', fontsize = 15)
plt.ylabel('data', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(sales))],
        ['baju', 'bajuu', 'topi', 'aksesoris'])

plt.title('SALES_DATA')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

barWidth = 0.5
fig = plt.subplots(figsize =(12, 8))
  
Chiko = [10]
Coa = [30]
Bros = [12]
Kokiaa = [11]

br1 = np.arange(len(Chiko))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
br4 = [x + barWidth for x in br3]

plt.bar(br1, Chiko, color ='r', width = barWidth,
        edgecolor ='grey', label ='Chiko')
plt.bar(br2, Coa, color ='g', width = barWidth,
        edgecolor ='grey', label ='Coa')
plt.bar(br3, Bros, color ='b', width = barWidth,
        edgecolor ='grey', label ='Bros')
plt.bar(br4, Kokiaa, color ='y', width = barWidth,
        edgecolor ='grey', label ='Kokiaa')
 
plt.xlabel('category', fontweight ='bold', fontsize = 15)
plt.ylabel('sales data', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(Chiko))],
        ['baju', 'bajuu', 'topi', 'aksesoris'])
 
plt.legend()
plt.show()
```
